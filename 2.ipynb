```python
import numpy as np
```

```python
### 1
def check_mult(A, B):
    error_message = "Error: Matrix product is undefined!"
    try:
        print("A*B shape: ", np.shape(A@B))
    except ValueError:
        print(error_message)
    try:
        print("B*A shape: ", np.shape(B@A))
    except ValueError:
        print(error_message)
```

```python
### 
A = np.ones((4, 2))
B = np.ones((4, 2))
check_mult(A, B)
```

```python
### 
A = np.ones((2, 5))
B = np.ones((5, 3))
check_mult(A, B)
```

```python
### 
A = np.ones((8, 3))
B = np.ones((3, 8))
check_mult(A, B)
```

```python
### 
A = np.ones((4, 4))
B = np.ones((4, 4))
check_mult(A, B)
```
