```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)
```

```python
### 1

A = np.array([[1, 2, 0],
              [0, 0, 5],
              [3, -4, 2],
              [1, 6, 5],
              [0, 1, 0]])

U, s, W = np.linalg.svd(A)
V = W.T
D = np.zeros_like(A, dtype=float)
D[np.diag_indices(min(A.shape))] = s
```

```python
print(f'Матрица U:\n{U}')
```

```python
print(f'Матрица D:\n{D}')
```

```python
print(f'Матрица V:\n{V}')
```

```python
### 2a

norm_e = D[0][0]
print(f'Евклидова норма:\n{norm_e}')

# numpy: 
print(np.linalg.norm(A, ord=2))
```

```python
### 2б

norm_f = np.sqrt(sum(D[i][i]**2 for i in range(np.linalg.matrix_rank(A))))
print(f'Норма Фробениуса:\n{norm_f}')

# numpy:
print(np.linalg.norm(A, ord='fro'))
```
